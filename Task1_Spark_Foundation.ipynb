Question: Predict the percentage of an student based on the no. of study hours.


What will be predicted score if a student studies for 9.25 hrs/ day?

```
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Reading dataset
url = "http://bit.ly/w-data"
dataset = pd.read_csv(url)
```

```
dataset.head() #displaying first 5 rows
```

```
#Preprocessing over the dataset
dataset.shape
```

```
dataset.info()
```

```
dataset.isnull().sum()  #checking for null values
```

```
#Plotting the distribution of scores
dataset.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours VS Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Scored')  
plt.show()
```

The above graph clearly states the positive inclination of hours studied to the percentage scored showing a linear relation.

```
#Splitting the dataset into train and test set
X = dataset.iloc[:, :-1].values  
Y = dataset.iloc[:, 1].values  
```

```
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```
#Training the dataset using Linear Regression model
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, Y_train) 
print("Training complete.")
```

```
print("Intercept \n",regressor.intercept_)
```

```
print("coefficient \n",regressor.coef_)
```

```
#Visualization of the regression

line = regressor.coef_*X+regressor.intercept_  #plotting regression line

#Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line);
plt.show()
```

Prediction of data for 9.25 hrs/day

Predicting the dataset

```
print(X_test) # Testing data - In Hours
Y_pred = regressor.predict(X_test) # Predicting the scores
```

```
dataset = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})  #Comparing the actual and predicted 
dataset
```

```
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred)) 
```

```
accuracy = regressor.score(X_test,Y_test)
print(accuracy*100,'%')
```

```
#Testing our own dataset
pred=regressor.predict(np.array([9.25]).reshape(1, 1))
print("The Predicted Score: ",pred)
```
